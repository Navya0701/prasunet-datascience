```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_355682.csv'
data = pd.read_csv(file_path, skiprows=4)
##here we are skipping the first 4 rows as the dataset contains some random metadata
```

```python
data.head()
```

```python
data.tail()
```

```python
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
##cleaning the dataframe containing random metadata
```

```python
years = data.columns[4:]
country_name = 'World'
population = data[data['Country Name'] == country_name].iloc[0, 4:]
```

```python
population.index = pd.to_numeric(population.index)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=population.index, y=population.values, hue=population.index, palette='viridis', legend=False)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Distribution Over Years')
plt.xticks(rotation=45)
plt.tight_layout()
```
